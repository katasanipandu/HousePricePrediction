```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
housing.keys()
```

```python
print(housing.DESCR)
```

```python
print(housing.data)
```

```python
print(housing.target)
print("="*25)
print(housing.feature_names)
```

```python
dataset = pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
dataset['Price']=housing.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset.corr()
```

```python
#Independent and Dependent features
X = dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
#Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
```

```python

regression=LinearRegression()
```

```python

regression.fit(X_train,y_train)
```

```python

## print the coefficients and the intercept
print(regression.coef_)
```

```python

print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python

### Prediction With Test Data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python

## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
# Draw straight line (ideal case)
plt.plot(y_test, y_test, color='red')
plt.scatter(y_test,reg_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")

plt.show()
```

```python

import pickle
#Pickling is the process of saving a Python object into a file so you can reuse it later.

#👉 It converts an object into a byte stream (serialization).
```

```python

pickle.dump(regression,open('regmodel.pkl','wb'))
```

```python

pickled_model=pickle.load(open('regmodel.pkl','rb'))
```

```python
pickled_model.predict(scaler.transform(housing.data[0].reshape(1, -1)))
```
